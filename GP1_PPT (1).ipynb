```python
import tkinter as tk
from tkinter import messagebox
import random

opciones = [
    "Piedra", "Papel", "Tijera", "Lagarto", "Spock",
    "Dragón", "Hielo", "Tornado", "Veneno", "Sol",
    "Luna", "Sombra", "Magia", "Bestia", "Cristal"
]

reglas = {
    "Piedra": ["Tijera", "Lagarto"],
    "Papel": ["Piedra", "Spock"],
    "Tijera": ["Papel", "Lagarto"],
    "Lagarto": ["Spock", "Papel"],
    "Spock": ["Tijera", "Piedra"],

    "Dragón": ["Aire", "Lagarto"],
    "Hielo": ["Agua", "Dragón"],
    "Tornado": ["Fuego", "Lagarto"],
    "Veneno": ["Dragón", "Spock"],
    "Sol": ["Agua", "Hielo"],
    "Luna": ["Dragón", "Rayo"],
    "Sombra": ["Papel", "Aire"],
    "Magia": ["Metal", "Veneno"],
    "Bestia": ["Lagarto", "Sombra"],
    "Cristal": ["Rayo", "Papel"]
}

rondas_totales = random.randint(3, 7)
ronda_actual = 1
puntos_jugador = 0
puntos_maquina = 0

def jugar(eleccion_jugador):
    global ronda_actual, puntos_jugador, puntos_maquina

    eleccion_maquina = random.choice(opciones)

    if eleccion_jugador == eleccion_maquina:
        resultado = "hubo empate"
    elif eleccion_maquina in reglas.get(eleccion_jugador, []):
        resultado = f"gano la ronda"
        puntos_jugador += 1
    else:
        resultado = f"perdio la ronda"
        puntos_maquina += 1

    messagebox.showinfo("resultado",
                        f"Ronda {ronda_actual} de {rondas_totales}\n"
                        f"Tú elegiste: {eleccion_jugador}\n"
                        f"Máquina eligió: {eleccion_maquina}\n\n"
                        f"{resultado}")

    ronda_actual += 1

    if ronda_actual > rondas_totales:
        mostrar_resultado_final()

def mostrar_resultado_final():
    if puntos_jugador > puntos_maquina:
        final = "gano la partida"
    elif puntos_jugador < puntos_maquina:
        final = "la maquina gano"
    else:
        final = "fin del juego =empate="

    messagebox.showinfo("Juego Terminado",
                        f"RESULTADOS FINALES\n\n"
                        f"Tus puntos: {puntos_jugador}\n"
                        f"Puntos de la máquina: {puntos_maquina}\n\n"
                        f"{final}")
    ventana.quit()


ventana = tk.Tk()
ventana.title("Mega Piedra, Papel, Tijera y Más")
ventana.geometry("500x700")
ventana.config(bg="lightyellow")


titulo = tk.Label(ventana, text="¡Juguemos Piedra, Papel, Tijera... y Más!",
                  font=("Arial", 18, "bold"), bg="lightyellow")
titulo.pack(pady=10)

subtitulo = tk.Label(ventana, text=f"La máquina decidió {rondas_totales} rondas",
                     font=("Arial", 14), bg="lightyellow")
subtitulo.pack(pady=5)


frame_botones = tk.Frame(ventana, bg="lightyellow")
frame_botones.pack(pady=20)

for opcion in opciones:
    boton = tk.Button(frame_botones, text=opcion,
                      font=("Arial", 12), width=20, height=2,
                      command=lambda o=opcion: jugar(o))
    boton.pack(pady=5)


ventana.mainloop()
```
