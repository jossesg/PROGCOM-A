```python
!pip install ipywidgets
from ipywidgets import Button, VBox, Output
import random
from IPython.display import display


opciones = [
    "Piedra", "Papel", "Tijera", "Spock", "Heavy", "Lagarto",
    "Pistola", "Rayo", "Dragón", "Agua", "Aire", "Esponja",
    "Lobo", "Árbol", "Humano", "Serpiente", "Fuego", "Diablo"
]


salida = Output()


def jugar(eleccion):
    pc = random.choice(opciones)
    if eleccion == pc:
        resultado = f"empate : {eleccion} | PC: {pc}"
    elif eleccion == "Diablo":
        resultado = f"ganaste : {eleccion} | PC: {pc}"
    else:

        if random.choice([True, False]):
            resultado = f"ganaste  : {eleccion} | PC: {pc}"
        else:
            resultado = f"perdiste  : {eleccion} | PC: {pc}"


    with salida:
        salida.clear_output()
        print(resultado)


botones = [Button( description=op ) for op in opciones]


for b in botones:
    b.on_click(lambda x, op=b.description: jugar(op))


display(VBox(botones))
display(salida)
```

```python
import random


opciones = [
    "Piedra", "Papel", "Tijera", "Spock", "Heavy", "Lagarto",
    "Pistola", "Rayo", "Dragón", "Agua", "Aire", "Esponja",
    "Lobo", "Árbol", "Humano", "Serpiente", "Fuego", "Diablo"
]

print("=== JUEGO PIEDRA, PAPEL, TIJERA Y MÁS ===")

rondas = int(input("Cuántas rondas quiere jugar? "))


puntos_jugador = 0
puntos_pc = 0

for ronda in range(1, rondas + 1):
    print(f" Ronda {ronda}")
    print("Opciones:")
    for i, op in enumerate(opciones, start=1):
        print(f"{i}. {op}")


    eleccion_num = int(input("elige el número de tu jugada: "))
    jugador = opciones[eleccion_num - 1]


    pc = random.choice(opciones)

    print(f"tu elegiste: {jugador}")
    print(f"la consola eligió: {pc}")


    if jugador == pc:
        print("hubo empate")
    elif jugador == "Diablo":
        print("ganaste!")
        puntos_jugador += 1
    else:
        if random.choice([True, False]):
            print("ganaste")
            puntos_jugador += 1
        else:
            print("perdiste")
            puntos_pc += 1


print(" RESULTADO FINAL ")
print(f"Tus puntos: {puntos_jugador}")
print(f"Puntos PC: {puntos_pc}")

if puntos_jugador > puntos_pc:
    print("gano el juego")
elif puntos_jugador < puntos_pc:
    print("la consola gano")
else:
    print("¡empataron")
```
